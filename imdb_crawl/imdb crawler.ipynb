```python
#This program receives the first 50 animations from the imdb site and displays information including
#title, links, genre, descriptions, duration, year and links 
#using the Pandas library.

from bs4 import BeautifulSoup
import requests
import pandas as pd

site = requests.get(r"https://www.imdb.com/search/title/?genres=animation&sort=num_votes,desc&explore=title_type,genres")
soup = BeautifulSoup(site.text , 'html.parser')

temp = soup.select('#main > div > div.lister.list.detail.sub-list > div > div > div.lister-item-image.float-left > a')

#link
link=[i['href']for i in temp]
links= ["https://www.imdb.com/"+i for i in link]


#title
temp = soup.select("#main > div > div.lister.list.detail.sub-list > div > div > div.lister-item-content > h3 > a")
title = [i.text for i in temp]


#descriptions
temp= soup.select('#main > div > div.lister.list.detail.sub-list > div.lister-list > div > div > p.text-muted')
description=[i.text for i in temp]
descriptions= description[1::2]


#duration
temp = soup.select('#main > div > div.lister.list.detail.sub-list > div > div > div.lister-item-content > p > span.runtime')
duration = [i.text for i in temp]


#genre
temp = soup.select('#main > div > div.lister.list.detail.sub-list > div > div > div.lister-item-content > p > span.genre')
genre = [i.text for i in temp]


#year
temp = soup.select('#main > div > div.lister.list.detail.sub-list > div > div > div.lister-item-content > h3 > span.lister-item-year.text-muted.unbold')
year = [i.text for i in temp]


data = pd.DataFrame({
    'title':title,
    'links':links,
    'genre':genre,
    'descriptions':descriptions,
    'duration':duration,
    'year':year
}
)

data


```
